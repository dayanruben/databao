```python
import logging

import pandas as pd

import databao

logging.basicConfig(level=logging.INFO)
```

```python
rfm = pd.read_csv("examples/data/rfm_segmented.csv")
rfm
```

```python
marketing = pd.read_csv("examples/data/segment_marketing_plan.csv")
marketing
```

```python
session = databao.open_session("my_session")
session.add_df(rfm)
session.add_df(marketing)
thread = session.thread()
```

```python
thread.ask("Which segments have the highest total revenue and what discount policy is currently assigned to each?").df()
```
